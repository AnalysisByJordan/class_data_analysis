```python
import pandas
import numpy
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
table = pandas.read_csv('data.csv')
table.shape
```

```python
pandas.set_option('display.max_columns', 200)
pandas.set_option('display.max_rows', 200)
```

```python
ID = [i for i in range(118)]
ID = list(map(str, ID)) 
ID = ["#" + i for i in ID]
table = table.drop(table.columns[[0]], axis=1)
table.insert(0, 'ID', ID ,)
Section = [1 if x=="North" else 2 for x in table.Section]
table = table.drop(table.columns[[1]], axis=1)
table.insert(1, 'Section', Section ,)
table
```

```python
norm = table.copy()
norm.iloc[:,5:] = ( table.iloc[:,5:] - table.mean()) / table.std()
x = norm.reindex(sorted(norm.columns), axis=1)
A = x.drop(x.columns[[13, 17, 20, 26, 32]], axis=1)
A
```

```python
zmax = norm.iloc[:,5:].abs().idxmax()

outliers = pandas.DataFrame(
    {
        'Column': zmax.index, 
        'ID': norm.ID.iloc[zmax],
        'Value': list( table.lookup( zmax, zmax.index ) )
    }
)
B = outliers.sort_values('Column')
B.Value = B.Value.astype(int)
B
```

```python
New = x
New = New.drop(New.columns[[13, 17, 20, 26]], axis=1)
cols = list(New.columns.values)
cols.pop(cols.index('Status'))
New = New[['Status'] + cols]
New
```

```python
fig = plt.figure( figsize=(40,10) )

C = pandas.plotting.parallel_coordinates( 
    New, # data frame
    'Status', # grouping variables (determining different colors of lines)
    New.columns.values[1:], # a list of x-axis column names (must be numeric) 
    colormap=plt.cm.rainbow # colors of the lines
)
C
```

```python
Outlierness = x
Outlierness = Outlierness.drop(Outlierness.columns[[13, 17, 20, 26, 32]], axis=1)
Outlierness = Outlierness.abs()
OS = Outlierness**2
```

```python
totals = numpy.sum(OS, axis = 1)
totalsroot = totals**.5
```

```python
IDS = table.ID.copy()

D = IDS.to_frame()
D.insert(1,'Outlierness',totalsroot)
D = D.sort_values('Outlierness', ascending=False)
D
```

```python
corr = A.corr()
img = matplotlib.pyplot.matshow(
    corr, 
    cmap=matplotlib.pyplot.cm.BrBG, 
    vmin=-1, 
    vmax=1 
)
img.figure.set_size_inches( 5, 5 )
img.axes.set_xticklabels(A.columns, rotation='vertical' )
img.axes.set_yticklabels(A.columns, rotation='horizontal' )
img.axes.set_xticks(range(A.columns.size))
img.axes.set_yticks(range(A.columns.size))
img.figure.colorbar( img )
E = img
E
```

```python
table5 = [(idx, col, corr[col].loc[idx])
           for idx in corr.index
           for col in corr.columns]
F = pandas.DataFrame(table5, columns=['Column_name1', 'Column_name2', 'Correlation_Value'])
F = F[F.Correlation_Value != 1]
F = F.drop_duplicates(['Correlation_Value'])
F = F.sort_values(['Correlation_Value'], ascending = False)
F
```
